```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
clusters_range = np.arange(1, 12, 1)
inertias = []
for n in clusters_range:
    model = KMeans(n_clusters=n).fit(df)
    inertias.append(model.inertia_)
```

```python
sns.lineplot(data=pd.DataFrame({"x": clusters_range, "y": inertias}), x="x", y="y")
```

```python
kmeans = KMeans(n_clusters=3).fit(df)
labels = kmeans.labels_
results_df = pd.concat([df, pd.Series(labels, name="Cluster")], axis=1)
```
